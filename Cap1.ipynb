```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(url)
df = df[0]
df.head(10)
```

```python
df = df.rename(columns = {'District':'Borough'})
df = df[df['Borough'] != 'Not assigned']
df = df.reset_index(drop = True)
df.head()
```

```python
import geocoder 
import folium
latlist = list()
longlist = list()
for i, j in zip(df['Postal Code'], df['Neighbourhood']):
    addy = '{}, Toronto, Ontario'.format(i)
    g = Nominatim(user_agent = 'foursquare_agent')
    loc = g.geocode(addy)
    try:
        latt = loc.latitude
        long = loc.longitude
        latlist.append(latt)
        longlist.append(long)
        print(latt, long)
    except:
        try:
            address = '{}, Toronto, Ontario'.format(j)
            j = Nominatim(user_agent = 'foursquare_agent')
            location = j.geocode(address)
            latitude = location.latitude
            longitude = location.longitude
            latlist.append(latitude)
            longlist.append(longitude)
        except:
            latlist.append('NA')
            longlist.append('NA')
latlngdf = pd.DataFrame({'Latitude':latlist,'Longitude':longlist})
df = pd.concat([df, latlngdf],axis = 1)
df.head(15)
```

```python
df = df[df['Latitude'] != 'NA']
df = df[df['Longitude'] != 'NA']
df.head(15)
```

```python
g = Nominatim(user_agent = 'foursquare_agent')
loc = g.geocode('Toronto, Canada')
lat_tor = loc.latitude
long_tor = loc.longitude
map_tor = folium.Map(location = [lat_tor,long_tor],zoom_start = 11)
for lat, long, label in zip(df['Latitude'],df['Longitude'],df['Postal Code']):
    label = folium.Popup(label, parse_html = True)
    folium.CircleMarker([lat,long],radius = 5,popup = label, color = 'blue',fill = True,fill_color = '#3186cc',fill_opacity = .7,parse_html = False).add_to(map_tor)
map_tor
```

```python
num_clus = 5 #depends on what seems correct
latlngdf = df.drop(['Postal Code','Neighbourhood','Borough'],axis = 1)
kmeans = KMeans(n_clusters = num_clus, random_state = 0).fit(latlngdf)
df.insert(0,'Cluster Labels',kmeans.labels_)
df.head(10)
```

```python
map_tor2 = folium.Map(location = [lat_tor,long_tor],zoom_start = 11)
colors = ['blue','red','green','yellow','orange','purple']
for lat, lng, clus, code in zip(df['Latitude'],df['Longitude'],df['Cluster Labels'],df['Postal Code']):
    color = colors[clus]
    label = '{} Cluster {}'.format(code, clus)
    folium.CircleMarker([lat,lng],radius = 5,popup = label,color = color, fill = False).add_to(map_tor2)
```
